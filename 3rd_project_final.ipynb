**Import dataset**

```python
db=pd.read_csv('/content/HR_data_week3_new_final.csv')
```

**Data Transformation**

```python
db.DOB.head(10)
```

```python
0db['DOB'] = pd.to_datetime(db['DOB'], errors='coerce')
current_year = datetime.today().year

# Adjust DOB by subtracting 100 years if the year is in the future
db['DOB'] = db['DOB'].apply(lambda d: d - pd.DateOffset(years=100) if pd.notnull(d) and d.year > current_year else d)
```

```python
# Today's date
today = pd.to_datetime('today')

# Calculate age correctly
db['Age'] = today.year - db['DOB'].dt.year
db['Age'] -= ((today.month < db['DOB'].dt.month) | ((today.month == db['DOB'].dt.month) & (today.day < db['DOB'].dt.day))).astype(int)
```

```python
db.drop(['MarriedID','MaritalStatusID','GenderID'],axis=1,inplace= True)
```

```python
db.shape
```

```python
db['DateofHire']=pd.to_datetime(db['DateofHire'])
db['DateofTermination']=pd.to_datetime(db['DateofTermination'])
```

```python

db['LastPerformanceReview_Date']=pd.to_datetime(db['LastPerformanceReview_Date'])
```

```python
db.fillna({'DateofTermination':0},inplace=True)
```

```python
db.isnull().sum()
```

```python
db.fillna({'ManagerID':'nil'},inplace=True)
```

```python
# Step 1: Split by comma into Last and the rest
db[['LastName', 'FirstMiddle']] = db['Employee_Name'].str.split(',', n=1, expand=True)

# Step 2: Extract First and Middle (if exists)
db[['FirstName', 'MiddleName']] = db['FirstMiddle'].str.strip().str.extract(r'(\w+)\s*(\w*)')

# Step 3: Combine all in correct order: First Middle Last
db['SortedName'] = db.apply(
    lambda row: ' '.join(filter(None, [row['FirstName'], row['MiddleName'], row['LastName']])),
    axis=1
)
```

```python
db.head()
```

```python
db.drop('Employee_Name', axis=1, inplace=True)
cols = ['SortedName'] + [col for col in db.columns if col != 'SortedName']
db = db[cols]
```

```python
db.drop(['FirstMiddle','LastName','FirstName','MiddleName'], axis=1, inplace=True)
```

```python
db.head(10)
```

```python
db.rename(columns={'SortedName': 'Employee_Name'}, inplace=True)
```

```python
db['ManagerID'] = db['ManagerID'].astype(object)
```

**Productivity rate**

```python
perf_map = {
    'Exceeds': 4,
    'Fully Meets': 3,
    'Needs Improvement': 2,
    'PIP': 1
}
db['PerformanceScoreNum'] = db['PerformanceScore'].map(perf_map)
```

```python
db['ProductivityRate'] = (
    (0.4 * db['PerformanceScoreNum']) +
    (0.3 * db['EngagementSurvey']) +
    (0.2 * db['EmpSatisfaction']) +
    (0.1 * np.log1p(db['SpecialProjectsCount']))
) / (1 + db['Absences'] + db['DaysLateLast30'])
```

```python
db.head()
```

```python
db['DateofHire'] = pd.to_datetime(db['DateofHire'])
db['DateofTermination'] = pd.to_datetime(db['DateofTermination'], errors='coerce', format='%m-%d-%Y')

# Fill missing termination dates (those still employed) with today's date
db['DateofTermination'] = db['DateofTermination'].fillna(pd.to_datetime('today'))
```

```python
db['EmploymentDuration_Days'] = (db['DateofTermination'] - db['DateofHire']).dt.days

# Optionally, calculate the gap in years
db['EmploymentDuration_Years'] = db['EmploymentDuration_Days'] / 365
```
